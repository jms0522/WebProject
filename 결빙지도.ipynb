```python
ice = pd.read_csv("13_22_freezing.csv", encoding = 'CP949')
```

```python
ice1 = ice.copy()
ice1
```

```python
ice_column = ['시도시군구명','지점명','사고건수','사상자수','사망자수','경도','위도']
```

```python
new_ice = ice1[ice_column]
```

```python
new_ice.count()
```

```python
seoul_data = new_ice[new_ice['시도시군구명'].str.contains('서울')]
seoul_data.count()
```

```python
from folium.plugins import MarkerCluster
# 지도 초기 설정
m = folium.Map(location=[36.5, 127.5], zoom_start=7)
# 초기 지도 위치와 줌 레벨 설정
marker_cluster = MarkerCluster().add_to(m)
# 데이터프레임 순회하며 마커 추가
for index, row in new_ice.iterrows():
    popup_text =  f'''
        <h3 style="color:black; font-size:16px;">{row['지점명']}</h3>
        <p style="font-size:12px;">사고건수: {row['사고건수']} <i class="fas fa-car-crash" style="color:red"></i></p>
        <p style="font-size:12px;">사상자수: {row['사상자수']} <i class="fas fa-user-injured" style="color:blue"></i></p>
    '''
    tooltip_text = f"지점명:{row['지점명']}"
    folium.Marker([row['위도'], row['경도']],tooltip =tooltip_text, popup=folium.Popup(popup_text, max_width=300),icon=folium.Icon('lightblue', icon='flash')).add_to(marker_cluster)

# 지도를 HTML 파일로 저장하여 시각화
m.save('결빙사고_지역_지도.html')
```
